```python
data_by_profess = data.groupby('Profession')
data_by_profess.size()
```

```python
# fill empty Birthyear
bym = data["BirthYear"].mean()
data["BirthYear"].fillna(bym, inplace = True)
```

```python
# fill empty Profession
data["Profession"].fillna('BDM', inplace = True)
```

```python
# drop the outlier from Amount col
data.drop(19)
```

```python
# drop Department column
data=data.drop(columns=['Department'])
```

```python
# fix data in consistently in profession column
data['Profession']= data['Profession'].replace('manager','Manager')
data['Profession']= data['Profession'].replace('hr','HR')
data['Profession']= data['Profession'].replace({'etudient':'Student', 'student':'Student'})
data['Profession']= data['Profession'].replace('bdm','BDM')
```

```python
data_by_cl = data.groupby('ClientID')
data_by_cl.size()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['Profession']=encoder.fit_transform(data['Profession'])
data['Risk']=encoder.fit_transform(data['Risk'])
data
```
