````python
url = 'https://www.emsc-csem.org/Earthquake/'
earthquakes = requests.get(url).content
soup = BeautifulSoup(earthquakes,"lxml")
earthqk = soup.find('tbody', {'id': 'tbody'}).find_all("tr");

latest = 20
latest_earth_list = [];

for earthquake in earthqk[:latest]:
    # Date and time
    date, time = earthquake.find('td', {'class': 'tabev6'}).find('a').text.split();
    # Latitude and longitude
    lat_deg, lon_deg = earthquake.find_all('td', {'class': 'tabev1'});
    lat_dir, lon_dir, magnitude = earthquake.find_all('td', {'class': 'tabev2'});
    lat_deg = f"{lat_deg.text.strip()} {lat_dir.text.strip()}";
    lon_deg = f"{lon_deg.text.strip()} {lon_dir.text.strip()}";
    # Region
    region = earthquake.find('td', {'class': 'tb_region'}).text.strip();
    # Create list of information and append
    earthquake_summary = [date, time, lat_deg , lon_deg, region];
    latest_earth_list.append(earthquake_summary);
    
df = pd.DataFrame(latest_earth_list, columns=['Date', 'Time', 'Latitude', 'Longitude', 'Region']);
df
for 
#header =soup.find_all('thead')[0]
#col= header.text
#etq = soup.find_all('tbody')[0]
#rows = etq.find_all('tr')
#rows = [row.text.strip().split("\n") for row in rows]
#print(rows)
#import pandas as pd
#colnames = col[0]
#data = rows[1:21]
#df = pd.DataFrame(data, columns=colnames)
#df[['Rank','Language','Speakers(millions)']]
#df[['Rank','Language','Speakers(millions)']]
#date = soup.find_all('td',{'class','tabev6'})
#latitude = soup.find_all('td',{'class','tabev1'})
#longitude = soup.find_all('td',{'class','tabev2'})
#region_name = soup.find_all('td',{'class','tb_region'})
#info = [element.info for element in soup.find_all(tags)]
#tags = ['a',tabev6','tabev1','tabev2','tb_region']
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
native_spk = requests.get(url).content
soup = BeautifulSoup(native_spk,"lxml")
languages = soup.find_all('table',{'class':'wikitable sortable'})[0]
rows = languages.find_all('tr')
rows = [row.text.strip().split("\n") for row in rows]
import pandas as pd
colnames = rows[0]
data = rows[1:11]
df = pd.DataFrame(data, columns=colnames)
df[['Rank','Language','Speakers(millions)']]
````
