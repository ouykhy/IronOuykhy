```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = pd.read_csv(r"C:\Users\ouykh\Ironhack\DAFT_NOV_21_01\module_2\Project_Week_4\diamonds_clean.csv")
diamonds = diamonds.drop('Unnamed: 0', 1)
```

### MATPLOTLIB

```python
diamonds.hist(column='carat',figsize=(7,4),color="r",linewidth=0.5, edgecolor="white")
plt.xlabel('Carat')
plt.ylabel('Number of diamonds')
plt.title('Distribution of diamonds by weight')

diamonds.plot.scatter(x='carat',y='price',figsize=(7,4),color="r")
plt.xlabel('Carat')
plt.ylabel('Price of diamond')
plt.title('Price of diamond by weight')

diamonds.hist(column='price',figsize=(7,4),color="b",linewidth=0.5, edgecolor="white")
plt.xlabel('Price')
plt.ylabel('Number of diamonds')
plt.title('Distribution of diamonds by price')

diamonds_depthmean = diamonds[["depth","price"]].groupby(["depth"]).agg("mean")
diamonds_depthmean.plot(color="purple")
plt.xlabel('Depth')
plt.ylabel('Mean diamond price')
plt.title('Mean diamond price by depth')

plt.show()
```

```python
diamonds["cut"].value_counts(normalize=True).plot(kind='pie')
plt.title('Distribution of diamonds by cut quality')

diamonds_cutmean = diamonds[["cut","price"]].groupby(["cut"]).agg("mean")
diamonds_cutmean.plot.barh()
plt.xlabel('Mean diamond price')
plt.ylabel('Cut quality')
plt.title('Mean diamond price by cut quality')

diamonds_tablemean = diamonds[["table","price"]].groupby(["table"]).agg("mean")
diamonds_tablemean.plot(color="green")
plt.xlabel('Table percentage - Width of top/widest point')
plt.ylabel('Mean diamond price')
plt.title('Mean diamond price by table')

plt.show()
```

```python
diamonds["table"].value_counts(normalize=True).plot(kind='pie')
plt.title('Distribution of diamonds by width of top of diamond')
plt.show()
```

### SEABORN

```python
plt.figure(figsize=(6,5), dpi= 80)
sns.kdeplot(diamonds['carat'], shade=True, color="g")
plt.title('Distribution of diamonds by weight')
plt.show()

sns.violinplot("price", data = diamonds)
plt.title('Distribution of diamonds by price',fontsize=15)
plt.show()
```

```python
sns.jointplot(x = 'carat', y = 'price', data = diamonds)
plt.title('Price of diamonds by weight',fontsize=15)
plt.show()

sns.countplot(x ='cut', data = diamonds, palette='rainbow')
plt.title('Distribution of diamonds by cut quality')

plt.show()
```

```python
sns.boxplot(x='cut',y='price',data=diamonds,palette='rainbow')
plt.title('Price of diamonds by cut quality')
plt.show()

sns.relplot(x='depth', y='price', hue="price",size="price", data=diamonds)
plt.title('Price of diamonds by depth %')
plt.show()

sns.distplot(diamonds['table'], bins = 30, kde = False, color = 'red')
plt.title('Distribution of diamonds by width of top relative to widest point',fontsize=15)
plt.show()

sns.set_theme(style="ticks")
sns.jointplot(x='table', y='price', kind="hex", color="#4CB391",data=diamonds)
plt.title('Price of diamonds by Table percentage',fontsize=15)
plt.show()
```
