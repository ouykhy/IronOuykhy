### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(movie_metadata.country, movie_metadata.movie_title.count())
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab(movie_metadata.country, [movie_metadata.director_name,movie_metadata.genres])
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(movie_metadata.color,'Average',values=movie_metadata.duration,aggfunc='mean')
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
pd.crosstab(movie_metadata.director_name,'Average',values=movie_metadata.director_facebook_likes,aggfunc='mean')
```
